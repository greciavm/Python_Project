```python
###### import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
Confirmed = "data/Confirmed.csv"
Deaths = "data/Deaths.csv"
Recovered = "data/Recovered.csv"

deaths= pd.read_csv(Deaths)
deathsByday=deaths.copy()
i=40
j=0
while (j<=104):
    while (i >=5):
        deathsByday.iloc[j,i]=deathsByday.iloc[j,i]-deathsByday.iloc[j,i-1]
        i = i - 1
    j=j+1
    i=40

deathsByday.head()

```
